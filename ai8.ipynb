```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
df = pd.read_csv("D:\CC GENERAL.csv")
df
```

```python
df=df.drop("CUST_ID", axis=1)
column_names = df.columns
for column in column_names:
    print(column + ' - ' + str(df[column].isnull().sum()))
```

```python
df['MINIMUM_PAYMENTS'].fillna(df['MINIMUM_PAYMENTS'].mean(skipna=True), inplace=True)
df['CREDIT_LIMIT'].fillna(df['CREDIT_LIMIT'].mean(skipna=True), inplace=True)
df
```

```python
from sklearn.cluster import AgglomerativeClustering
model=AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='complete')
x = df[['ONEOFF_PURCHASES_FREQUENCY','PURCHASES_FREQUENCY']]
y = model.fit_predict(x)
#clust_labels=model.fit_predict(df)
```

```python
agglomerative=pd.DataFrame(y)
agglomerative
```

```python
import matplotlib.pyplot as plt
fig =plt.figure()
ax = fig.add_subplot(111)
scatter = ax.scatter (df ["PURCHASES_FREQUENCY"] , df ["ONEOFF_PURCHASES_FREQUENCY"] , c= agglomerative[0], s=50)
ax.set_title("Agglomerative Clustering")
ax.set_xlabel("PURCHASES_FREQUENCY (k$)")
ax.set_ylabel("ONEOFF_PURCHASES_FREQUENCY")
plt.colorbar(scatter)
```

```python
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10,7))
plt.title("Customer Dendrograms")
dend=shc.dendrogram(shc.linkage(df, method="complete"))
dend
```

```python
from sklearn.cluster import KMeans  
kmeans=KMeans(n_clusters=4, random_state=0)
x = df[['PURCHASES_FREQUENCY','ONEOFF_PURCHASES_FREQUENCY']]
y = kmeans.fit(x)
```

```python
plt.scatter(df["PURCHASES_FREQUENCY"][y.labels_ == 0],          
            df["ONEOFF_PURCHASES_FREQUENCY"][y.labels_ == 0],s=80,c='magenta',label='Careful')
plt.scatter(df["PURCHASES_FREQUENCY"][y.labels_ == 1],
           df["ONEOFF_PURCHASES_FREQUENCY"][y.labels_ == 1],s=80,c='yellow',label='Standard')
plt.scatter(df["PURCHASES_FREQUENCY"][y.labels_ == 2],
           df["ONEOFF_PURCHASES_FREQUENCY"][y.labels_ == 2],s=80,c='green',label='Target')
plt.scatter(df["PURCHASES_FREQUENCY"][y.labels_ == 3], 
           df["ONEOFF_PURCHASES_FREQUENCY"][y.labels_ == 3],s=80,c='cyan',label='Careless')
plt.scatter(df["PURCHASES_FREQUENCY"][y.labels_ == 4], 
           df["ONEOFF_PURCHASES_FREQUENCY"][y.labels_ == 4],s=80,c='burlywood',label='Sensible')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', label = 'Centroids')
plt.title('Clusters of Customers')
plt.xlabel('PURCHASES_FREQUENCY')
plt.ylabel('ONEOFF_PURCHASES_FREQUENCY')
plt.legend()
plt.show()
```

```python
sum_of_sqrdis = []
k = range(1,10)
for k in k :
        km = KMeans(n_clusters=k)
        km = km.fit(df)
        sum_of_sqrdis.append(km.inertia_)
plt.plot(sum_of_sqrdis, "ro-")
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3).fit_transform(df)
fig = plt.figure(figsize=(12, 7), dpi=80, facecolor='w', edgecolor='k')
ax = plt.axes(projection="3d")
ax.scatter3D(pca.T[0],pca.T[1],pca.T[2],c=kmeans.labels_,cmap='Spectral')
xLabel = ax.set_xlabel('X')
yLabel = ax.set_ylabel('Y')
zLabel = ax.set_zlabel('Z')
```

```python
from sklearn.metrics import silhouette_score
from sklearn.metrics import calinski_harabasz_score
from sklearn.metrics import davies_bouldin_score
def metrics(data,labels):
    ss = silhouette_score(data,labels)
    chs = calinski_harabasz_score(data,labels)
    dbs = davies_bouldin_score(data,labels)
    return ss , chs ,dbs
```

```python
ag_ss,ag_chs,ag_dbs=metrics(x,agglomerative)
```

```python
km_ss,km_chs,km_dbs=metrics(x,y.labels_)
```

```python
results={"Agglomerative Clustering":{"silhouette_score":ag_ss,"calinski_harabasz_score":ag_chs,"davies_bouldin_score":ag_dbs},"K-means Clustering":{"silhouette_score":km_ss,"calinski_harabasz_score":km_chs,"davies_bouldin_score":km_dbs}}
results = pd.DataFrame(results)
results.transpose()
```
